# Frequency of Events

```python
import numpy as np
n = 10000 # 模拟次数
a = np.random.rand(n) # 生成[0,1)的随机数
p = np.array([0.2, 0.05, 0.01, 0.06, 0.08, 0.1, 0.3, 0.05, 0.03, 0.12])
# 事件发生概率分布
cp = np.cumsum(p) # 对事件发生概率逐次累加
c = [np.sum(a <= cp[0])] # 事件发生的频数A1 
for i in range(1, len(p)):
    c.append(np.sum((a > cp[i - 1]) & (a <= cp[i])))
print(c) # 输出结果
```

# Estimate Pi

```python
import numpy as np
import matplotlib.pyplot as plt
N = 10000
x,y = np.random.uniform(-1,1,size=(2,N)) # 生成个随机点N
inside = (x**2+y**2)<=1
mpi = inside.sum()*4/N # 求的近似值pi 
error = abs((mpi-np.pi)/np.pi)*100
# 计算理论值与模拟值的误差
outside = np.invert(inside) 
plt.plot(x[inside],y[inside],'b.') 
plt.plot(x[outside],y[outside],'r.') 
plt.axis('square')
plt.show()
print('%.3f'% mpi,'%.3f%%'% error)
```

# Find Maximum Value

```python
import numpy as np
from matplotlib.pyplot import plot, show
from scipy.optimize import fminbound
# 绘制f(x)函数图像
f =lambda x: (1-x**3)*np.sin(3*x)
x0 = np.linspace(-2*np.pi,2*np.pi,100) 
plot(x0,f(x0))
show()
# 基于函数求解fminboundf(x)最大值
x1 = fminbound(lambda x: -f(x),-2*np.pi,2*np.pi) 
y1 = f(x1)
print([x1,y1])
```

As we can see, if we directly compute the maximum height of the function using gradients, it can get stuck in a local maximum. Here, we can use Monte Carlo Simulation to find the global maximum value of the function.

```python
# 利用随机模拟的方法求解f(x)最大值
x2 = np.random.uniform(-2*np.pi,2*np.pi,100) 
y2 = f(x2)
y2_max = y2.max()
x2_max = x2[y2==y2_max] 
print([x2_max,y2_max])
```

# Maximum Profit

```python
from scipy.stats import poisson 
a=2
b=3
lamda = 10
p = 1-a/b
u = poisson.ppf(p,lamda) # 求最佳订购量
p1 = poisson.cdf(u-1,lamda) # 验证最佳购进量 
p2 = poisson.cdf(u,lamda) 
print([u,p1,p,p2])
```
