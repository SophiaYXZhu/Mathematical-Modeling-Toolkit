# Linear Programming Case Study
Variables: <br/>
• si :第i种投资项目，如股票、债券 <br/>
• ri , pi , qi :分别为si 的平均收益率，交易费率，风险损失率 <br/>
• ui : si的交易定额 <br/>
• r0 :同期银行利率 <br/>
• xi : 投资项目si的资金 <br/>
• a : 投资风险度 <br/>
• Q:总体收益 <br/>
Assumptions: <br/>
• 投资数额M相当大，为了便于计算，假设M = 1 <br/>
• 投资越分散，总风险越小，总体风险用投资项目si中最大的一个 风险来度量 <br/>
• n种资产si之间是相互独立的 <br/>
• 在投资的这一时期内，ri , pi , qi , r0 为定值 <br/>
Objective Functions: <br/>
• min max{qi * xi|i = 1, 2, ...n} <br/>
• max sum(ri − pi) * xi <br/>

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as op #导入相关库 
a = 0 #定义的取值a
profit_list = [] #记录最大收益
a_list = [] #记录的取值a
while a<0.05:
    x1=(0,None) #定义决策变量取值范围 
    c=np.array([-0.05,-0.27,-0.19,-0.185,-0.185]) #定义目标函数系数
    A = np.hstack((np.zeros((4,1)), np.diag([0.025,0.015,0.055,0.026]))) #定义不等式约束条件左边系数
    b=a*np.ones((4,1)) #定义不等式约束条件右边系数 
    Aeq=np.array([[1,1.01,1.02,1.045,1.065]]) #定义等式约束条件左边系数
    beq=np.array([1]) #定义等式约束条件右边系数
    res=op.linprog(c,A,b,Aeq,beq,bounds=(x1,x1,x1,x1,x1))
    profit = -res.fun
    profit_list.append(profit)
    a_list.append(a)
    a += 0.001
plt.figure(figsize=(10,7))
plt.plot(a_list,profit_list)
plt.xlabel('a')
plt.ylabel('Profit') #绘制风险偏好与最大收益的曲线图a
```
