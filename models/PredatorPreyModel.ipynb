# Lotka Volterra Model

```python
import matplotlib
import matplotlib.pyplot as plt 
plt.style.use('ggplot')

r = 1
d = 0.5
a = 0.1
b = 0.02
x0 = 5
y0 = 12
deltaT = 0.01
TotTime = 30
timeStep = TotTime/deltaT
xlist = []
ylist = []
xlist.append(x0) # x0
ylist.append(y0) # y0
TimeList = [i*deltaT for i in range(int(timeStep))]
for time in TimeList:
    xlist.append(xlist[-1] +deltaT * (r *xlist[-1] - a*xlist[-1]*ylist[-1] ) )
    ylist.append(ylist[-1] +deltaT * (-d *ylist[-1] + b*xlist[-1]*ylist[-1] ) )
# 
plt.figure(figsize = (20,5)) 
plt.plot(TimeList,xlist[:-1],label = 'Small Fish') 
plt.plot(TimeList,ylist[:-1],label = 'Large Fish') 
plt.legend()
```

The model is periodic, as shown in the graph below (that is closed):

```python
plt.plot(xlist,ylist)
```

The next cell computes the mean of the prey and the predator population. The result aligns with what we observe in the graph above by drawing a vertical and horizontal line across the minimum and maximum points of x and y respectively.

```python
import numpy as np
x_bar = d/b
y_bar = r/a
print(x_bar)
print(y_bar)
```

# Lotka Volterra Model With Logistic Term
In real life, we rarely observe a predator-prey model that cycles periodically forever.
To simulate the effect of a naturally normalizing phenomenon, we add an additional logistic term to the prey and population's rates of growth.

```python
r1 = 0.2
r2 = 0.1
N1 = 50
N2 = 60
sigma1 = 1.2
sigma2 = 2.5
x0 = 25
y0 = 2
deltaT = 0.01
TotTime = 50
timeStep = TotTime/deltaT
xlist = []
ylist = []
xlist.append(x0) # x0
ylist.append(y0) # y0
TimeList = [i*deltaT for i in range(int(timeStep))]
for time in TimeList:
    xlist.append(xlist[-1] +deltaT * (r *xlist[-1]) *(1 - xlist[-1]/N1 - sigma1 * ylist[-1]/N2))
    ylist.append(ylist[-1] +deltaT * (r *ylist[-1]) *(-1 +sigma2* xlist[-1]/N1 - ylist[-1]/N2))
#  
plt.figure(figsize = (20,5)) 
plt.plot(TimeList,xlist[:-1]) 
plt.plot(TimeList,ylist[:-1])
```
